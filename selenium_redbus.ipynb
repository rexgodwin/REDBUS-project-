# kerala routes

```python
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 20)

def Kerala_link_route(path):   
    LINKS_KERALA = []
    ROUTE_KERALA = []
    
    # Scrape the first page
    paths = driver.find_elements(By.XPATH, path)
    print(f"Step 1: Found {len(paths)} elements on the first page")
    
    for links in paths:
        d = links.get_attribute("href")
        LINKS_KERALA.append(d)
        
    for route in paths:
        ROUTE_KERALA.append(route.text)
    
    # Pagination loop
    for i in range(2, 4):  # Start from the second page
        try:
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i}]')
            
            # Scroll to the next button
            driver.execute_script("arguments[0].scrollIntoView(true);", next_button)
            time.sleep(1)  # Adjust sleep time if necessary
            
            # Wait for the next button to be clickable
            wait.until(EC.element_to_be_clickable((By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i}]')))
            
            # Click the next button using JavaScript
            driver.execute_script("arguments[0].click();", next_button)
            
            # Scrape the next page
            paths = driver.find_elements(By.XPATH, path)
            print(f"Step {i}: Found {len(paths)} elements on page {i}")
            
            for links in paths:
                d = links.get_attribute("href")
                LINKS_KERALA.append(d)
                
            for route in paths:
                ROUTE_KERALA.append(route.text)
            
        except NoSuchElementException:
            print(f"No more pages to paginate at step {i}")
            break
        except TimeoutException:
            print(f"Timeout while waiting for pagination element at step {i}")
            break
        except ElementClickInterceptedException:
            print(f"Element click intercepted at step {i}")
            break
            
    return LINKS_KERALA, ROUTE_KERALA

# Example usage
driver.get("https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile")
LINKS_KERALA, ROUTE_KERALA = Kerala_link_route("//a[@class='route']")

# Debug: Print the results
print("Links:", LINKS_KERALA)
print("Routes:", ROUTE_KERALA)

# Close the driver
driver.quit()
```

```python
df_k=pd.DataFrame({"Route_name":ROUTE_KERALA,"Route_link":LINKS_KERALA})
df_k
```

```python
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 20)

def WB_link_route(path):   
    LINKS_WB = []
    ROUTE_WB = []
    
    # Scrape the first page
    paths = driver.find_elements(By.XPATH, path)
    print(f"Step 1: Found {len(paths)} elements on the first page")
    
    for links in paths:
        d = links.get_attribute("href")
        LINKS_WB.append(d)
        
    for route in paths:
        ROUTE_WB.append(route.text)
    
    # Pagination loop
    for i in range(2, 4):  # Start from the second page
        try:
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i}]')
            
            # Scroll to the next button
            driver.execute_script("arguments[0].scrollIntoView(true);", next_button)
            time.sleep(1)  # Adjust sleep time if necessary
            
            # Wait for the next button to be clickable
            wait.until(EC.element_to_be_clickable((By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i}]')))
            
            # Click the next button using JavaScript
            driver.execute_script("arguments[0].click();", next_button)
            
            # Scrape the next page
            paths = driver.find_elements(By.XPATH, path)
            print(f"Step {i}: Found {len(paths)} elements on page {i}")
            
            for links in paths:
                d = links.get_attribute("href")
                LINKS_WB.append(d)
                
            for route in paths:
                ROUTE_WB.append(route.text)
            
        except NoSuchElementException:
            print(f"No more pages to paginate at step {i}")
            break
        except TimeoutException:
            print(f"Timeout while waiting for pagination element at step {i}")
            break
        except ElementClickInterceptedException:
            print(f"Element click intercepted at step {i}")
            break
            
    return LINKS_WB, ROUTE_WB

# Example usage
driver.get("https://www.redbus.in/online-booking/wbtc-ctc/?utm_source=rtchometile")
LINKS_WB, ROUTE_WB = WB_link_route("//a[@class='route']")

# Debug: Print the results
print("Links:", LINKS_WB)
print("Routes:", ROUTE_WB)

# Close the driver
driver.quit()
```
